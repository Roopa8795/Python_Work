# Task8:Performance Categorization (Data Manipulation)

```python
def performance_category(score):
    if score>=250:
        return "Excellent"
    elif score>=200:
        return "good"
    else:
        return "Needs improvement"
df["Performance Level"]=df["Total"].apply(performance_category)
df["Performance Level"]
```

```python
df["Performance Level"].value_counts()
```
